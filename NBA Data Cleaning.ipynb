```python
import pandas as pd
import numpy as np
```

```python
#To load NBA players and teams data set.
NBA_Players= pd.read_csv("players.csv")
NBA_Teams= pd.read_csv("NBA_teams.csv")
```

# Cleaning Scraped Player Data

```python
#To view the first 30 players info
NBA_Players.head(30)
```

```python
#Deleting the unnamed and RK column
del NBA_Players["Unnamed: 0"]
del NBA_Players["Rk"]
```

```python
#To view the first 30 players info
NBA_Players.head(30)
```

```python
#Remove the '*' in some of the player names from the Player column
NBA_Players["Player"] = NBA_Players["Player"].str.replace("*", "", regex=False)
```

```python
#To view the first 30 players info
NBA_Players.head(30)
```

```python
#To focus on players total stats and last team they played for by year so as to avoid duplicating players that played for multiple teams in a year.
def single_row(stat):
    if stat.shape[0]==1:
        return stat
    else:
        One_row = stat[stat["Tm"]=="TOT"]
        One_row["Tm"] = stat.iloc[-1, :]["Tm"]
        return One_row
NBA_Players = NBA_Players.groupby(["Player", "Year"]).apply(single_row)
```

```python
#To view the first 30 players info
NBA_Players.head(30)
```

```python
#To remove the two additional indexes (Player and year) created from merging the player stats into one row per year
NBA_Players.index= NBA_Players.index.droplevel()
NBA_Players.index= NBA_Players.index.droplevel()
```

```python
#To view the first 30 players info
NBA_Players.head(30)
```

```python
#To view the NaN values in the data set
NBA_Players.isna().sum()
```

# Cleaning the Scraped Teams Data

```python
#To view the first 20 rows of the data set
NBA_Teams.head(20)
```

```python
#To remove the rows which have the word "division" included
NBA_Teams= NBA_Teams[~NBA_Teams["W"].str.contains("Division")]
```

```python
#To view the first 20 rows of the data set
NBA_Teams.head(20)
```

```python
#Remove the '*' in some of Team names from the Team column
NBA_Teams["Team"] = NBA_Teams["Team"].str.replace("*", "", regex=False)
```

```python
#To view the first 20 rows of the data set
NBA_Teams.head(20)
```

```python
#Deleting the unnamed column
del NBA_Teams["Unnamed: 0"]
```

```python
#To view the first 20 row of the data set
NBA_Teams.head(20)
```

# Merging player data with Team Data 

```python
#Creating a column (Team) to assist in mapping between the two data frames
NBA_Team_Abb={}
#Uploading csv file with team names and abbreviations
with open("NBA_Team_Abb.csv") as f:
    lines = f.readlines()
    for line in lines[1:]:
        abbrev,name=line.replace("\n", "").split(",")
        NBA_Team_Abb[abbrev] = name
```

```python
NBA_Players["Team"]= NBA_Players["Tm"].map(NBA_Team_Abb)
NBA_Players
```

```python
#Merging the player data set with the team data set
NBA_Player_Stats= NBA_Players.merge(NBA_Teams, how='outer', on=["Team", "Year"] )
```
