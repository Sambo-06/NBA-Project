```python
#Most 3 point shot made by players who played more than 60 games (1991 - 2023
higest_3pm = df[df["G"]> 60].sort_values("3P", ascending=False).head(10)
higest_3pm.plot.bar("Player", "3P")
plt.show()
```

```python
# Most 3 point shot made each year by players who played more than 65 games (1991 - 2023)
highest_3p_peryear= df.groupby("Year").apply(lambda x: x.sort_values("3P", ascending=False).head(1))
highest_3p_peryear.plot.bar("Year", "3P")
plt.show()
```

```python
#Highest 3 point percentage 65 games and attempted atleast three 3 point shots a game
filtered_df= df[(df["3PA"]> 3) & (df["G"]> 65)]
highest_3per = filtered_df.sort_values("3P%", ascending=False).head(10)
highest_3per.plot.bar("Player", "3P%")
```

```python
#To show the distrubtion of 3 point percetage from 1991 to 2023
df["3P%"].plot(kind="hist", bins=20)
plt.show()
```

Histogram shows that a large number of players have 0% percent three point percentage most likely because they never attempeted a three point shot or missed every three point attempt they attempted. Three point percentages higher that 60% most likey have fewer three point attempts and/ or played less than 70 games.

# Step 3: Feature Relationships

```python
#To see what values correlate with the 3P% column
df_corr= df.corr()["3P%"]
print(df_corr)
```

```python
#Bar chart showing correlation
df_corr.plot.bar()
plt.title('Correlation of "3P%" to features')
plt.xlabel('Features')
plt.ylabel('Correlation')
plt.show()
```

```python
#Heatmap showing correlation
correlations = df.corr()['3P%']
plt.figure(figsize=(30, 20))
correlation_df = pd.DataFrame(correlations)
sns.heatmap(correlation_df.transpose().round(2), annot=True, cmap='coolwarm', vmin=-1, vmax=1,  annot_kws={"size": 15})
plt.title('Correlation Heatmap for 3P% Column')
plt.xticks(fontsize=25) 
plt.yticks(fontsize=25)
plt.show()
```

From the heatmap above, we see that the features with the stongest correlations to the 3 point percentage are 3 point attempts and 3 points made. Other features that correlate are field goal attempts points game, assists, and blocks.
